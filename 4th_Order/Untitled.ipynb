```python
import numpy as np
import matplotlib.pyplot as plt
import os
font = {'size'   : 18}
axes = {'linewidth':2,'labelpad':8.0,}
ticks={'major.size':7,'major.width':2}
plt.rc('font', **font)
plt.rc('axes',**axes)
plt.rc(['xtick','ytick'],**ticks)
```

```python
B=['Sun','Mercury','Venus','Earth','Mars','Jupiter','Saturn','Uranus','Neptune']
#dE=np.loadtxt('dE.txt')
year=86400*365.25
data=[]
for i in os.listdir('Data/'):
    if('pos' in i):
        data.append(np.loadtxt('Data/'+i))
dt=np.loadtxt('Data/dt.txt')
```

```python
print(*np.unique(dt))
```

```python
fig,ax=plt.subplots(figsize=(30,10))
ax.plot(dt,'o')
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(15,15))
for i in range(len(data)):
    ax.plot(data[i][:,0],data[i][:,1],label=B[i])
#ax.set_xlim(-2,2)
#ax.set_ylim(-2,2)

ax.legend()
plt.show()      
```
